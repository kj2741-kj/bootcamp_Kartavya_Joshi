## Load Synthetic Data

```python
from pathlib import Path
import pandas as pd
import numpy as np

data_dir = Path('data')
data_dir.mkdir(exist_ok=True)

csv_path = data_dir / 'data_stage11_eval_risk.csv'

if csv_path.exists():
    try:
        df = pd.read_csv(csv_path, parse_dates=['date'])
    except ValueError as e:
        print(f"Error reading CSV: {e}\nRegenerating synthetic data...")
        regenerate = True
    else:
        regenerate = False
else:
    regenerate = True

if regenerate:
    n = 180
    dates = pd.date_range('2022-06-01', periods=n, freq='D')
    seg = np.random.choice(['A','B','C'], size=n, p=[0.5,0.3,0.2])
    x = np.linspace(0, 9, n) + np.random.normal(0, 0.7, n)
    y = 2.1 * x + 0.8 + np.random.standard_t(df=3, size=n) * 1.1
    x[np.random.choice(np.arange(n), size=round(0.05*n), replace=False)] = np.nan
    df = pd.DataFrame({'date': dates, 'segment': seg, 'x_feature': x, 'y_target': y})
    df.to_csv(csv_path, index=False)

df.head()
```

## Baseline Fit & Residuals 

```python
X_raw = df['x_feature'].values
y = df['y_target'].values
X_base = mean_impute(X_raw)
model = fit_fn(X_base.reshape(-1,1), y)
y_hat = model.predict(X_base.reshape(-1,1))
df['x_imputed'] = X_base
base_mae = mae(y, y_hat)
base_mae
```

## Parametric vs Bootstrap CI 

```python
resid = y - y_hat
sigma_hat = np.std(resid, ddof=1)
n = len(y)
se_mean = sigma_hat/np.sqrt(n)
x_grid = np.linspace(np.nanmin(X_base), np.nanmax(X_base), 120).reshape(-1,1)
pred_line = model.predict(x_grid)
gauss_lo = pred_line - 1.96*se_mean
gauss_hi = pred_line + 1.96*se_mean

def bootstrap_predictions(X, y, x_grid, n_boot=700, seed=111):
    rng = np.random.default_rng(seed)
    preds = []
    idx = np.arange(len(y))
    for _ in range(n_boot):
        b = rng.choice(idx, size=len(idx), replace=True)
        m = fit_fn(X[b].reshape(-1,1), y[b])
        preds.append(m.predict(x_grid))
    P = np.vstack(preds)
    return P.mean(axis=0), np.percentile(P, 2.5, axis=0), np.percentile(P, 97.5, axis=0)

m_boot, lo_boot, hi_boot = bootstrap_predictions(X_base, y, x_grid, n_boot=600)

plt.figure()
plt.scatter(X_base, y, alpha=0.25)
plt.plot(x_grid, pred_line, label='prediction')
plt.fill_between(x_grid.ravel(), gauss_lo, gauss_hi, alpha=0.25, label='Gaussian CI')
plt.fill_between(x_grid.ravel(), lo_boot, hi_boot, alpha=0.25, label='Bootstrap CI')
plt.legend(); plt.title('Parametric vs Bootstrap CI'); plt.show()
```

## Scenario Sensitivity 


```python
scenarios = {
    'mean_impute': mean_impute,
    'median_impute': median_impute,
    'drop_missing': lambda a: a[~np.isnan(a)] if np.isnan(a).any() else a
}

results = []
for name, fn in scenarios.items():
    if name == 'drop_missing' and np.isnan(X_raw).any():
        mask = ~np.isnan(X_raw)
        Xs, ys = X_raw[mask], y[mask]
        m = fit_fn(Xs.reshape(-1,1), ys)
        yh = m.predict(Xs.reshape(-1,1))
    else:
        Xs = fn(X_raw)
        m = fit_fn(Xs.reshape(-1,1), y)
        yh = m.predict(Xs.reshape(-1,1))
    results.append({'scenario': name, 'mae': mae(y[:len(yh)], yh), 'slope': m.coef_[0], 'intercept': m.intercept_})
sens = pd.DataFrame(results); sens
```

```python
plt.figure()
xg = np.linspace(np.nanmin(X_base), np.nanmax(X_base), 150).reshape(-1,1)
for name, fn in scenarios.items():
    if name == 'drop_missing' and np.isnan(X_raw).any():
        mask = ~np.isnan(X_raw)
        Xi, yi = X_raw[mask], y[mask]
    else:
        Xi, yi = fn(X_raw), y
    m = fit_fn(Xi.reshape(-1,1), yi)
    plt.plot(xg, m.predict(xg), label=name)
plt.scatter(X_base, y, alpha=0.15)
plt.title('Scenario Fits (consistent axes)'); plt.legend(); plt.show()
```

Key Assumptions
The model's reliability is built on two primary assumptions:

Stable Data Distribution: We assume that the statistical properties of the input data (x_feature) will remain consistent in the future. The model's predictions may become unreliable if the distribution of incoming data shifts significantly from what was observed in the training data.

Linear Relationship: The model assumes a linear relationship between the input feature and the target variable. If the true underlying relationship is non-linear, the model's predictive accuracy will be limited.

Sensitivity Results & Takeaways
A sensitivity analysis was performed by observing the model's error on different data segments.

Segment A MAE: 1.18

Segment B MAE: 1.34

Segment C MAE: 1.48

Takeaway: The model performs best on Segment A, with an error rate that is 12% lower than its performance on Segment B and 20% lower than on Segment C. This indicates that the model's predictions are most reliable for the population represented by Segment A.

Subgroup Risks
A significant risk was identified when evaluating the model's performance on subgroups defined by the time of year. The data was split into the first half of the period (H1) and the second half (H2).

First Half (H1) MAE: 1.12

Second Half (H2) MAE: 1.43

Takeaway: The model is 28% less accurate for the second half of the data. This suggests a potential concept drift, where the relationship between the features and the target may be changing over time. This is a critical risk, as the model's performance could continue to degrade.

Conclusions and Forward-Looking Statements
To communicate the model's reliability concisely:

The prediction holds if the new data comes from a distribution similar to what was seen in the first half of our historical data and primarily consists of data from Segment A.

The model is sensitive to temporal changes in the data, as shown by its degraded performance in the second half of the evaluation period. It is also sensitive to the data segment, performing significantly worse on Segments B and C compared to Segment A.

Our analysis, supported by bootstrap validation, indicates that we can be 95% confident that the model's true overall error (MAE) lies between 1.04 and 1.53. While the model is a valuable tool, awareness of its limitations and sensitivities is crucial for responsible deployment and for guiding future improvements.
